# Write a Python Program to simulate a simple banking system, where users can `create accounts`, `deposit money`, `withdraw money` and `check balance` using Python data structures and functions

## Creating the Accounts - Dictionary | so that we can store multiple information in one place

```python
from faker import Faker
```

```python
fake_data_generator = Faker('en_IN')
```

```python
accounts = []
```

```python
def create_account(holder_name, initial_balance, age, account_type):
    global accounts
    accounts.append(
        {
            "account_no": fake_data_generator.random_number(digits=16,fix_len=True),
            "holder_name": holder_name,
            "balance": initial_balance,
            "address": fake_data_generator.address(),
            "age": age,
            "type": account_type
        }
    )

def account_details_user_input():
    holder_name = input("Enter the holder full name:")
    initial_balance = float(input("Enter the initial balance:"))
    age = int(input('Enter the age of the holder:'))
    account_type = input("Enter the type of account:")
    create_account(holder_name, initial_balance, age, account_type)
```

```python
def withdraw_money(account_no,txn_amount):
    global accounts
    for account in accounts:
        if account['account_no'] == account_no:
            if account['balance'] >= txn_amount:
                account['balance'] -= txn_amount
            else:
                print('\nInsufficient Balance')

def withdraw_money_user_input():
    account_no = int(input('Please enter the Account No. for withdrawal:'))
    txn_amount = float(input('Enter Amount:'))
    withdraw_money(account_no,txn_amount)  
```

```python
def deposit_money(account_no,txn_amount):
    global accounts
    for account in accounts:
        if account['account_no'] == account_no:
            account['balance'] += txn_amount

def deposit_money_user_input():
    account_no = int(input('Please enter the Account No. for deposit:'))
    txn_amount = float(input('Enter Amount:'))
    deposit_money(account_no,txn_amount)    
```

```python
def check_balance(account_no):
    for account in accounts:
        if account['account_no'] == account_no:
            print('Balance:',account['balance'])

def check_balance_user_input():
    account_no = int(input('Please enter the Account No. you want to check balance for:'))
    check_balance(account_no) 
```

```python
print("------ Simple Banking System ------")
print("Welcome to our bank, we hope to server you the best")
print("Please choose from the below menu options:")
print("1. Create Account")
print("2. Display Accounts")
print("3. Withdraw Money")
print("4. Deposit Money")
print("5. Check Balance")
print("6. Exit")

while True:
    menu_selection = int(input('\nEnter your choice:'))
    if menu_selection == 1:
        account_details_user_input()
    elif menu_selection == 2:
        print(accounts)
    elif menu_selection == 3:
        withdraw_money_user_input()
    elif menu_selection == 4:
        deposit_money_user_input()
    elif menu_selection == 5:
        check_balance_user_input()
    elif menu_selection == 6:
        break
    else:
        print('Invalid choice, please try again')
        
print('\nThank you for choosing our banking services. We hope to serve you better in the future.')
```
