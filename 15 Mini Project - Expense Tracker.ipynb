# Build a Simple Expense Tracker where users can record(create), manage(update, delete) and list their daily expenses.

```python
from faker import Faker
fake_data_generator = Faker('en_IN')
```

```python
expenses = []
```

```python
def add_expense():
    global expenses
    
    exp_name = input('Enter the Expense Name:')
    exp_category = input('Enter the Expense Category:')
    exp_date = input('Enter the Expense Date(YYYY-MM-DD):')            
    exp_amount = float(input('Enter the Expense Amount:'))

    expenses.append(
        {
            "id": fake_data_generator.random_number(digits=5,fix_len=True),
            "name": exp_name,
            "category": exp_category,
            "date": exp_date,
            "amount": exp_amount
        }
    )
```

```python
def display_all_expenses():
    print(expenses)
```

```python
def update_expense():
    global expenses

    exp_id = int(input('Enter the Expense ID to be updated:'))
    
    print('Select from the below choices:')
    print('1. Name')
    print('2. Category')
    print('3. Date')
    print('4. Amount')
    print('5. Exit')

    while True:
        update_choice = int(input('Enter your choice:'))
        if update_choice == 1:
            exp_name = input('Enter the Expense Name:')
            for expense in expenses:
                if expense['id'] == exp_id:
                    expense['name'] = exp_name
        elif update_choice == 2:
            exp_category = input('Enter the Expense Category:')
            for expense in expenses:
                if expense['id'] == exp_id:
                    expense['category'] = exp_category
        elif update_choice == 3:
            exp_date = input('Enter the Expense Date(YYYY-MM-DD):')    
            for expense in expenses:
                if expense['id'] == exp_id:
                    expense['date'] = exp_date
        elif update_choice == 4:
            exp_amount = float(input('Enter the Expense Amount:'))
            for expense in expenses:
                if expense['id'] == exp_id:
                    expense['amount'] = exp_amount
        elif update_choice == 5:
            break
        else:
            print('Invalid Menu selection. Please try again!')
```

```python
def delete_expense():
    global expenses

    exp_id = int(input('Enter the Expense ID to be deleted:'))

    for i in range(len(expenses)):
        if expenses[i]['id'] == exp_id:
            del expenses[i]
```

```python
print('------ Simple Expense Tracker------')
print('Select from the below menu options:')
print('1. Add Expense')
print('2. Display All Expenses')
print('3. Update Expense')
print('4. Delete Expense')
print('5. Exit')

while True:
    choice = int(input('\nEnter your choice:'))
    if choice == 1:
        add_expense()
    elif choice == 2:
        display_all_expenses()
    elif choice == 3:
        update_expense()
    elif choice == 4:
        delete_expense()
    elif choice == 5:
        break
    else:
        print('Invalid Menu selection. Please try again!')
```
